```python
import regex
from string import punctuation
from nltk.tokenize import sent_tokenize
import numpy as np
```

```python
tolstoy = open('anna_karenina.txt').read()
```

```python
punctuation+= '–'
print(punctuation)
```

```python
def normalize(text):
    res = regex.findall('[а-я]+|[а-я]*?-[а-я]*', text.lower(), flags=regex.POSIX)
    return(res)
```

```python
sentences_tolstoy = [['<start>', '<start>'] + normalize(text) + ['<end>', '<end>'] for text in sent_tokenize(tolstoy)]
```

```python
def ngrammer(tokens, n):
    ngrams = []
    for i in range(0,len(tokens)-n+1):
        ngrams.append(' '.join(tokens[i:i+n]))
    return ngrams
```

```python
from collections import Counter
unigrams_tolstoy = Counter()
bigrams_tolstoy = Counter()
trigrams_tolstoy = Counter()

for sentence in sentences_tolstoy:
    unigrams_tolstoy.update(sentence)
    bigrams_tolstoy.update(ngrammer(sentence, n =2))
    trigrams_tolstoy.update(ngrammer(sentence, n = 3))
```

```python
matrix_tolstoy = np.zeros((len(bigrams_tolstoy), 
                   len(unigrams_tolstoy)))

id2word_tolstoy = list(unigrams_tolstoy)
word2id_tolstoy = {word:i for i, word in enumerate(id2word_tolstoy)}

id2bigram_tolstoy = list(bigrams_tolstoy)
bigram2id_tolstoy = {bigram:i for i, bigram in enumerate(id2bigram_tolstoy)}


#вероятность увидеть слово из униграммы после биграммы: 

for trigram in trigrams_tolstoy:
    word1, word2, word = trigram.split()
    bigram = ''.join(word1+' '+word2)
    matrix_tolstoy[bigram2id_tolstoy[bigram]][word2id_tolstoy[word]] = (trigrams_tolstoy[trigram]/bigrams_tolstoy[bigram])
```

```python
def generate(matrix_tolstoy, id2word_tolstoy, word2id_tolstoy, bigram2id_tolstoy,id2bigram_tolstoy, n=100, start='<start> <start>'):
    text = []
    current_idx = bigram2id_tolstoy[ start]

    for i in range(100):

        chosen = np.random.choice(matrix_tolstoy.shape[1], p=matrix_tolstoy[current_idx])

        text.append(id2word_tolstoy[chosen])



        if id2word_tolstoy[chosen] == '<end>':
            chosen = word2id_tolstoy['<start>']
            current_idx = bigram2id_tolstoy['<start> <start>']
        else:
            word1, word2 = id2bigram_tolstoy[current_idx].split()
            current_idx = bigram2id_tolstoy[''.join(word2 + ' ' + str(id2word_tolstoy[chosen]))]
    res = ' '.join(text)
    return res
```

```python
generate(matrix_tolstoy,id2word_tolstoy, word2id_tolstoy,  bigram2id_tolstoy, id2bigram_tolstoy)
```

* threshold  – Represent a score threshold for forming the phrases (higher means fewer phrases). A phrase of words a followed by b is accepted if the score of the phrase is greater than threshold. Heavily depends on concrete scoring-function, see the scoring parameter.)
